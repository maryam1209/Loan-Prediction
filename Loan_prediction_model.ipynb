3. Modeling & Evaluation

```python
#importing liabraries
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score,precision_score, recall_score
from sklearn.model_selection import GridSearchCV
from sklearn.inspection import PartialDependenceDisplay
```

```python
selected_features_for_training=file[["CreditScore","RiskScore","TotalDebtToIncomeRatio","CreditCardUtilizationRate","BankruptcyHistory","PreviousLoanDefaults","PaymentHistory","LengthOfCreditHistory","AnnualIncome","JobTenure","NetWorth","EmploymentStatus","EducationLevel","Age","HomeOwnershipStatus","SavingsAccountBalance"]]
```

```python
obj_col=selected_features_for_training.select_dtypes(include="O")
file_object_encoded=pd.get_dummies(obj_col,drop_first=True)
num_col=selected_features_for_training.select_dtypes(exclude="O")
file_num_scaled = pd.DataFrame(StandardScaler().fit_transform(num_col), columns=num_col.columns,index=num_col.index)
file_model_training = pd.concat([file_num_scaled, file_object_encoded], axis=1)
```

```python
file_model_training.info()
```

```python
x=file_model_training
y=file["LoanApproved"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

KNN

```python
knn=KNeighborsClassifier()
param_grid = {'n_neighbors': [3, 5, 7, 9, 11, 15],'weights': ['uniform', 'distance'],'metric': ['euclidean', 'manhattan']}
grid_knn = GridSearchCV(knn,param_grid,cv=5,scoring='roc_auc',verbose=1,n_jobs=-1)
grid_knn.fit(x_train, y_train)
best_knn=grid_knn.best_estimator_
y_pred_knn= best_knn.predict(x_test)
y_proba_knn= best_knn.predict_proba(x_test)[:, 1]
knn_auc = roc_auc_score(y_test,y_proba_knn)
knn_prec = precision_score(y_test,best_knn.predict(x_test))
knn_rec = recall_score(y_test, best_knn.predict(x_test))
print("---CONFUSION MATRIX---")
cm = confusion_matrix(y_test, y_pred_knn)
print(cm)
print("---CLASSIFICATION MATRIX---")
print(classification_report(y_test, y_pred_knn))
print(f"Best Cross Validation score of ROC-AUC: {grid_knn.best_score_:.4f}")
print(f"Test Set ROC-AUC: {knn_auc:.4f}")
```

Decision Tree

```python
param_grid_dt = {'max_depth': [3, 5, 10, 15, None],'min_samples_split': [2, 5, 10],'min_samples_leaf': [1, 2, 4],'criterion': ['gini', 'entropy']}
grid_search_dt = GridSearchCV(DecisionTreeClassifier(random_state=42),param_grid_dt,cv=5,scoring='roc_auc',n_jobs=-1)
grid_search_dt.fit(x_train, y_train)
best_tree_model = grid_search_dt.best_estimator_
y_pred_dt = best_tree_model.predict(x_test)
y_proba_dt = best_tree_model.predict_proba(x_test)[:, 1]
dt_auc = roc_auc_score(y_test,y_proba_dt)
dt_prec = precision_score(y_test, best_tree_model.predict(x_test))
dt_rec = recall_score(y_test, best_tree_model.predict(x_test))
print("CONFUSION MATRIX:")
cm = confusion_matrix(y_test, y_pred_dt)
cm_df = pd.DataFrame(cm, index=['Actual Reject', 'Actual Approve'],columns=['Predicted Reject', 'Predicted Approve'])
print(cm_df)
print("Classifaction matrix: ",classification_report(y_test, y_pred_dt))
print(f"Best Cross Validation score of ROC-AUC: {grid_search_dt.best_score_:.4f}")
print(f"Test Set ROC-AUC: {dt_auc:.4f}")
```

One tree-based model (Random Forest)

```python
param_grid_rf = {'n_estimators': [50, 100, 200],'max_depth': [5, 10, None],'criterion': ['gini', 'entropy'],'max_features': ['sqrt', 'log2']}
grid_rf = GridSearchCV(RandomForestClassifier(random_state=42),param_grid_rf,cv=5,scoring='roc_auc',n_jobs=-1,verbose=1)
grid_rf.fit(x_train, y_train)
best_rf = grid_rf.best_estimator_
y_pred_rf = best_rf.predict(x_test)
y_proba_rf = best_rf.predict_proba(x_test)[:, 1]
rf_auc = roc_auc_score(y_test,y_proba_rf)
rf_prec = precision_score(y_test, best_rf.predict(x_test))
rf_rec = recall_score(y_test, best_rf.predict(x_test))
print("CONFUSION MATRIX:")
cm_rf = confusion_matrix(y_test, y_pred_rf)
cm_rf_df = pd.DataFrame(cm_rf, index=['Actual Reject', 'Actual Approve'],columns=['Predicted Reject', 'Predicted Approve'])
print(cm_rf_df)
print("\nCLASSIFICATION MATRIX:")
print(classification_report(y_test, y_pred_rf))
print(f"Best Cross Validation scorem ROC-AUC: {grid_rf.best_score_:.4f}")
print(f"Test Set ROC-AUC: {rf_auc:.4f}")
```

```python
comparison_data={'Model': ['KNN', 'Decision Tree', 'Random Forest'],'ROC-AUC Score': [knn_auc, dt_auc, rf_auc],'Precision': [knn_prec, dt_prec, rf_prec],'Recall': [knn_rec, dt_rec, rf_rec]}
comparison= pd.DataFrame(comparison_data)
print("---MODEL COMPARISON TABLE---")
print(comparison.sort_values(by="ROC-AUC Score", ascending=False).reset_index(drop=True))
```

4. Model Interpretability

```python
importance=best_rf.feature_importances_
feature_names=x_train.columns
feature_importance_file=pd.DataFrame({'Feature': feature_names, 'Importance': importance})
feature_importance_df=feature_importance_file.sort_values(by='Importance',ascending=False)
plt.figure(figsize=(10, 6))
plt.barh(feature_importance_df['Feature'][:10], feature_importance_df['Importance'][:10])
plt.xlabel('Importance Score')
plt.title('Top 10 Features Influencing Loan Approval (Random Forest)')
plt.show()
```

```python
features_to_plot = feature_importance_df['Feature'][:3].tolist()
fig, ax = plt.subplots(figsize=(15, 5))
PartialDependenceDisplay.from_estimator(best_rf, x_train, features_to_plot, ax=ax)
plt.suptitle('Partial Dependencies Plot: Impact of Key Features on Approval')
plt.show()
```
